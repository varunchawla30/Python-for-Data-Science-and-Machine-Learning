```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
type(data)
```

```python
len(data)
```

The make_blobs() function returns a tuple of size 2. The first element is the data which is basically in form of numpy array.The second element contains the cluster values to which the each of the record belongs to.

```python
data[0].shape
```

```python
data[1].shape
```

```python
data[1]
```

```python
# Now, let's visualize our data using scatterplot
```

```python
plt.scatter(data[0][:,0], data[0][:,1],c=data[1])
```

#### Let's create clusters using K Means Clsutering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# Let's compare the clusters made by K Means algorithm with Original Clusters
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)


ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
# We can also try this K Means using n_clusters = 2
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```
