## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

The above command allows us to see the plots inside Jupyter Notebook. If we aren't using Jupyter Notebook then we've to use plt.show() at the end of each plot to get it displayed.

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

There are two ways to create a Matplotlib plot. One is Functional and the other one is Object Oriented.

```python
# Functional Method
```

```python
plt.plot(x,y)
```

```python
# If we use plt.show() at the end of it, we won't get Output cell. It just prints the plot like print() function.
plt.plot(x,y)
plt.show()
```

```python
# To add x label, y label and title to your plot, use the below commands:

plt.plot(x,y)
plt.xlabel('Number')
plt.ylabel('Number ** 2')
plt.title('Number & its Square')
```

Let's now create multi plots on the same canvas

```python
plt.subplot(1,2,1) # 1st argument is num of rows, 2nd argument is num of columns, 3rd argument is plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
```

```python
my_figure = plt.figure()

axes = my_figure.add_axes([0.1,0.1,0.8,0.8]) # [left,bottom,width,height]

axes.plot(x,y)
axes.set_xlabel('Num')
axes.set_ylabel('Num**2')
axes.set_title('My Plot')
```

```python
# Now, let's put two sets of figures on one canvas
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

Let's understand the above method
1. First, we created a figure.
2. Then, we manually created both the axes.
3. Finally, we plotted the graph using the axes which we created 

 #### Now, let's see how we can plot subplots using Object Oriented Method

```python
a = [1,2,3,4,5]
b = [num**3 for num in a]
```

```python
a
```

```python
b
```

```python
newFig,newAxes = plt.subplots(nrows=1,ncols=2) # By default, nrows=1 & ncols=1

newAxes[0].plot(a,b)
newAxes[0].set_title('First Plot')

newAxes[1].plot(b,a)
newAxes[1].set_title('Second Plot')
plt.tight_layout()
```

Basically, plt.subplots() creates a figure and automaically adds axes to them based on number of rows and columns we want.

```python
# newAxes is basically an array of matplotlib axes
newAxes
```

```python
# We can also use loop to iterate over axes array object as:

fi,axes_array = plt.subplots(nrows=2,ncols=1)

i=0
while i<len(axes_array):
    axes_array[i].plot(a,b)
    if i == 0:
        break
        
# We can also use for each loop for iterating
```

```python
# Let's take about legends now. Using legends, we can specify which plot denotes what
```

```python
twoInOneFig = plt.figure()
twoInOneAxes = twoInOneFig.add_axes([0,0,1,1])

twoInOneAxes.plot(x,x**2,label='X Squared')
twoInOneAxes.plot(x,x**3,label='X Cubed')

twoInOneAxes.legend() # It checks for each of the plot calls and sees if there's any label
```

### Color, Line Width and Line Style of Plots

```python
# We can also use RGB HexCodes for color argument
# Line width is the thickness of line
# alpha is tranparency of line
# We can use 'lw' and 'ls' as aliases for line width and line style respectively
# To mark the data points on plot, we use 'marker' argument and for its size, we use 'markersize'
# We can set limit of x axis and y axis using set_xlim() & set_ylim() respectively

aFig = plt.figure()
axis = aFig.add_axes([0,0,1,1])
axis.plot(x,y,color='red',linewidth=1.5,alpha=0.75,linestyle='-',marker='o',markersize=12,markerfacecolor='yellow',
          markeredgewidth=3,markeredgecolor='blue')
```
