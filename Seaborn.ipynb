### Distribution Plots

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # It's a DataFrame
```

```python
tips.head()
```

```python
len(tips)
```

#### 1. Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

#### 2. Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

In between the distplots, we have a scatter plot. It shows that as the total_bill increases, tip also increases

```python
# Now, let's see the argument 'kind' which we can provide to jointplot(). By default, it is scatter.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

It shows hexagonal representation of data points. Hexagons with more data points are darker & ones with less data points are lighter.

```python
# Let's see another example of kind using value as 'reg' i.e, regression
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# Let's see another example of kind using value as 'kde' 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

#### 3. Pair Plot

Pair Plot help us to visualize data easily by giving us plots between each possible combination of numerical columns in the data.

```python
sns.pairplot(tips)
```

Let's see the argument hue in which we can pass any categorical column as value to differentiate data distribution across different categories

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

#### 4. Rug Plot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

The main difference between rugplot and distplot is that instead of displaying histograms by counting how many dashes are there in a bin and showing it as number, rugplot only display data points as dashes along the distribution line.

KDE Plots stand for Kernel Density Estimation Plot. KDE is basically the sum of all the normal distributions around a Rugplot.

```python
sns.kdeplot(tips['total_bill'])
```

### Categorical Plots

```python
tips.head()
```

#### 1. Bar Plot

Bar Plot shows us the mean value of a numerical column vs any categorical column. It's because the default value of estimator argument in barplot() is mean function. The estimator argument takes an aggregate function as value. It's kind of a group-by visualization of data based on a categorical and numerical column.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

#### 2. Count Plot

It basically counts the number of occurences of values of a categorical column in the dataset.

```python
sns.countplot(x='sex',data=tips)
```

#### 3. Box Plot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary i.e., minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can tell you about your outliers and what their values are.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

#### 4. Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

#### 5. Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

#### 6. Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Swarm and Violin Plot together
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

#### 7. Factor Plot

You can also use factor plot to plot any type of plot by passing its name as value of argument kind.

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

### Matrix Plots

```python
tips.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

#### 1. Heat Map

In order for the heat map to work properly, our data should be in matrix form i.e., the index names and the column names should be there so that they give us a correalation between them. Indexes should be labels and not numerical. There are 2 ways to get it, that is either by pivot table or by correlation data.

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
# To get the actual value of correlation between different columns, pass the argument annot=True
```

```python
sns.heatmap(tc,annot=True)
```

```python
flights.head()
```

```python
# Now, let's convert the flights data to matrix form using pivot table
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp)
```

```python
# To differentiate between different values, we use linecolor and linewidth arguments. 
sns.heatmap(fp,linecolor='white',linewidth=1)
```

```python
# Using different cmap
sns.heatmap(fp,linecolor='white',linewidth=1,cmap='coolwarm')
```

#### 2. Cluster Map

A cluster map will cluster things in order so that similar groups are close to each other or we can say that groups with higher values are shown together and vice-versa. In this case, we can say that similar months and similar years are put together.

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

### Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# Now, using PairGrid mechanism of Seaborn, we can customize this PairPlot
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
tg = sns.FacetGrid(data=tips,col='time',row='smoker')
tg.map(sns.distplot,'total_bill')
```

```python
tg = sns.FacetGrid(data=tips,col='time',row='smoker')
tg.map(plt.scatter,'total_bill','tip')
```
